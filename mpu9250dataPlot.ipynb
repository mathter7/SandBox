```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("log2.csv",delimiter=",")
```

```python
data
```

```python
acc_x = data[:,0]
acc_y = data[:,1]
acc_z = data[:,2]
gyr_x = data[:,3]
gyr_y = data[:,4]
gyr_z = data[:,5]
mag_x = data[:,6]
mag_y = data[:,7]
mag_z = data[:,8]
```

```python
acc_x.size
```

```python
x = np.arange(acc_x.size)
```

```python
# 図のサイズ指定
plt.figure(figsize=(20,10))

plt.subplot(3,1,1)
plt.title("acc")
plt.plot(x,acc_x, label="x")
plt.plot(x,acc_y, label="y")
plt.plot(x,acc_z, label="z")
plt.legend()

plt.subplot(3,1,2)
plt.title("gyr")
plt.plot(x,gyr_x, label="x")
plt.plot(x,gyr_y, label="y")
plt.plot(x,gyr_z, label="z")
plt.legend()

plt.subplot(3,1,3)
plt.title("mag")
plt.plot(x,mag_x, label="x")
plt.plot(x,mag_y, label="y")
plt.plot(x,mag_z, label="z")
plt.legend()

plt.tight_layout() 
```
